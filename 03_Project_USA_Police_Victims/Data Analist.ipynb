```python
# create data
size_of_groups=df.manner_of_death.value_counts()

# Create a pieplot
plt.pie(size_of_groups,
       startangle = 90)

# add a circle at the center to transform it in a donut chart
my_circle=plt.Circle((0,0), 0.55, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
data = df.manner_of_death.value_counts()
colors = ('#6FC0F2','#F7D754')

# Creating plot
fig, ax = plt.subplots(figsize =(10, 5))

# Plot Pie
ax.pie(data, startangle = 90, colors = colors)

# Adding legend
legend = ['Shot','Shot and Tasered']
ax.legend(legend,
          title ="Legend:",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))


# add a circle at the center to transform it in a donut chart
my_circle=plt.Circle((0,0), 0.55, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

# Add Title
ax.set_title("My Title")
```

```python
sns.set_theme(style="whitegrid")
ax = sns.lineplot(data=ref, x=ref['0'],y=ref['Percent_completed_hs'])
ax.set(xlabel='SEX', ylabel='Count', title='USA Percentual completed high school Overall')

mean_line = ref.Percent_completed_hs.mean()
plt.axhline(y=mean_line, xmin=0.0, xmax=1.0, color='blue')

plt.gcf().set_size_inches(15, 8)
plt.xticks(rotation=90, fontsize = 10)
# plt.savefig("./Data/Percentual completed high school Overall.png", dpi=1600)
```

```python
sns.set_theme(style="whitegrid")

to_convert = ref['Median_income']
data = pd.to_numeric(to_convert, errors='coerce')

mean_line = data.mean()
plt.axhline(y=mean_line, xmin=0.0, xmax=1.0, color='blue')

ax = sns.lineplot(data=ref, x=ref['0'],y=data, color='red')
ax.set(xlabel='SEX', ylabel='Count', title='USA Median Income Overall')
plt.gcf().set_size_inches(15, 8)
plt.xticks(rotation=90, fontsize = 10)
# plt.savefig("./Img/USA Median_income Overall.png", dpi=1600)
```
