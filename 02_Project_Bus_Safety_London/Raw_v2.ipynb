## Creazione dataframe per il lavoro

```python
df = raw.copy()
```

```python
df.columns
```

```python
df.columns = ['Year', 'Month', 'Route', 'Operator', 'Group_Name',
       'Bus_Garage', 'Borough', 'Injury_Result_Description',
       'Incident_Event_Type', 'Victim_Category', 'Victims_Sex', 'Victims_Age']
```

### Eliminare da colonna Date
m = df.Month.str.split(pat="-")

```python
m = df.Month.str.split(pat="-")
m[0][1]
```

```python
df = df.replace(df['Month'],m[0][1])
df
```

```python
import numpy as np
m = df['Month'].astype(str)
month = np.array([Month[5:7] for Month in m])
month
```

```python
df.insert(2,'month', month)
```

```python
df.head(2)
```

```python
df.drop('Month', axis=1, inplace=True)
```

```python
df
```

replace mesi 

```python
annual_dict = {'01': 'Jan','02': 'Feb','03': 'Mar','04': 'Apr','05': 'May','06': 'Jun','07': 'Jul','08': 'Aug','09': 'Sep','10': 'Oct','11': 'Nov','12': 'Dec'}

df.month.replace(annual_dict, inplace=True)
```

```python
df
```

```python
df.Route.replace('(blank)', np.nan, inplace=True)
```

```python
df.Bus_Garage.replace('Garage Not Available', np.nan, inplace=True)
```
